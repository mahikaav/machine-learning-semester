## Built-in Methods

### 1) arange

The function numpy.arange() creates a one-dimensional array of type numpy.ndarray. The array contains values that are equally spaced within a given range. We can choose which parameters to mention depending on our requirements. 

1. start: The interval starts from this value, which is included in the output. If we don’t specify it, it defaults to 0

2. stop: The interval ends at this value, which is excluded from the output, unless step is not a whole number and the output length is affected by rounding errors.

3. step: The output values are separated by this amount. For any output out, this is the difference between out[i+1] and out[i]. If you don’t specify it, it defaults to 1. If you specify it as a positional argument, you must also specify start.

3. dtype: The output array has this data type. If we don’t specify it, it is inferred from the other arguments.

```python
np.arange(2,18) #starts from 2, 18 is not included
```

```python
np.arange(10,21,2) #if we want to print even values between 9 and 20 (incl) then we have to start with 10 because initial is included
```

```python
np.arange(16) #if no start is given, it starts from 0 as default
```

### 3) linspace

numpy.linspace() creates an array of equally spaced values between start and stop. It takes the following parameters:

1. start: the first value in the array (default 0)
2. stop: the last value in the array (unless endpoint is False)
3. num: the number of values in the array (default 50)
4. endpoint: whether to include stop in the array (default True)
5. retstep: whether to return the spacing between values (default False)
6. dtype: the data type of the array (default None)
It returns an array of the values and optionally the step size.

```python
np.linspace(1,10,7) #here 7 represents the number of divisions or elements we need
```

1. R: the number of rows in the output

3. k: the index of the diagonal (0 for main, positive for upper, negative for lower)
4. dtype: the data type of the output (default float)
It returns an array of shape (R, C) with ones on the k-th diagonal and zeros elsewhere.

### Broadcasting

Broadcasting is a feature of numpy arrays that allows them to perform arithmetic operations with arrays of different shapes. For example, we can add a scalar value to an array, or multiply two arrays with different dimensions, as long as they have compatible shapes.

```python
#Setting a value with index range aka Broadcasting
ranarr[0:5]=10
ranarr
```

I have broadcasted a scalar value (10) to a subarray of shape (5,) and replaced the original values with 10 so 10 is assigned to the first five elements of the array ranarr i.e., from indices 0 to 4.

### Indexing a 2D array

To access a two-dimensional array, we can use either arr_2d[row][col] or arr_2d[row,col]. The comma notation is clearer and more readable.

```python
twoD_arr = np.array(([1,5,10],[2,6,12],[3,9,12]))
twoD_arr
```

```python
twoD_arr[0]
```

To access the first row of a two-dimensional array in Python, we can use the index 0 inside square brackets. 

```python
twoD_arr[1]
```

```python
twoD_arr[1][0]
```

To get an individual element value from a two-dimensional array, we can use the format twoD_arr[row][col], where row and col are the indices of the row and column we want to access. The indices start from 0, so arr_2d[1][0] means the second row and the first column of the array.

```python
twoD_arr[1,2] #to get 12
```

```python
twoD_arr.shape
```

```python
twoD_arr[:2,1:]
```

[:2, 1:] means the first two rows and the last two columns of the array. We can also use a single colon (:) to indicate all the elements in that dimension. For example, twoD_arr [:, :3] means all the rows and the first three columns of the array.

You can also use negative indices to count from the end of the array. For example, twoD_arr [-1, -2] means the last row and the second last column of the array.

You can also use a step value to skip some elements along each dimension. For example, twoD_arr[::2, ::2] means every other row and every other column of the array.

```python
twoD_arr[:,:3]
```

```python
twoD_arr[:1,1:] #first row and last two columns
```

```python
twoD_arr[0:1,:] #means the first row and all the columns of the array
```

```python
twoD_arr[0:2,0:2] #means first 2 rows and first two columns
```

```python
twoD_arr[2,:] #last row, all columns
```

```python
twoD_arr[:,2] #all the rows and the third column of the original array 
```

arr_2d[:][2] means the third row of that copy, which is a one-dimensional array.

```python
twoD_arr[:][2]
```

twoD_arr[:] returns a copy of the original array, which is still a two-dimensional array. So, twoD_arr[:][2] means the third row of that copy.
